```python
from scipy.stats import pearsonr
import pandas as pd
import matplotlib.pyplot as plt
```

DIPK, LCO. 

```python

dataset = []
for i in range(5):
    current_dataset = pd.read_csv(f"./Data/DIPK/no_randomization/predictions_split_{i}.csv")
    dataset.append(current_dataset)

dataset = pd.concat(dataset, ignore_index=True)
MSE = (sum((dataset.loc[:,"predictions"] - dataset.loc[:,"response"])**2))/len(dataset)
pearson = pearsonr(dataset.loc[:,"predictions"], dataset.loc[:,"response"]).statistic

MSE = round(MSE, 3)
pearson = round(pearson, 3)


plt.text(-7, 9, f"DIPK \nMSE: {MSE} \nPearson: {pearson}", fontsize = 10, c='r')
plt.xlabel("Predicted Values")
plt.ylabel("Ground Truth Values")
plt.scatter(dataset.loc[:,"predictions"], dataset.loc[:,"response"], s=2)
plt.show()
```
